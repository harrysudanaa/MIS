```python
print("Ini konflik ges")
# Mencari min dan max latitude longitude
import rasterio
from rasterio.warp import transform_bounds
import pyproj

# Open the Landsat 8 image
image_path = "/content/drive/MyDrive/Landsat8/B4/LC08_L2SP_122065_20180706_20200831_02_T1_SR_B4.TIF"  # Replace with the actual path to your Landsat 8 image
dataset = rasterio.open(image_path)

# Get the coordinate reference system (CRS) information
crs = dataset.crs

# Get the bounds of the image in the CRS
bounds = dataset.bounds

# Define the source and target CRS
source_crs = pyproj.CRS.from_string(crs.to_string())
target_crs = pyproj.CRS.from_epsg(4326)  # EPSG 4326 is the coordinate system for latitude and longitude

# Transform the image bounds to latitude and longitude
transformer = pyproj.Transformer.from_crs(source_crs, target_crs, always_xy=True)
minx, miny = transformer.transform(bounds.left, bounds.bottom)
maxx, maxy = transformer.transform(bounds.right, bounds.top)

print(f"Min Longitude: {minx}")
print(f"Min Latitude: {miny}")
print(f"Max Longitude: {maxx}")
print(f"Max Latitude: {maxy}")
```

```python
# mencari koordinat di setiap pixel B4
import rasterio
import numpy as np
import pyproj

# Open the Landsat 8 image
image_path = "/content/drive/MyDrive/Landsat8/B4/LC08_L2SP_122065_20180706_20200831_02_T1_SR_B4.TIF"  # Replace with the actual path to your Landsat 8 image
dataset = rasterio.open(image_path)

# Get the coordinate reference system (CRS) information
crs = dataset.crs

# Define the source and target CRS
source_crs = pyproj.CRS.from_string(crs.to_string())
target_crs = pyproj.CRS.from_epsg(4326)  # EPSG 4326 is the coordinate system for latitude and longitude

# Get the transformation function
transformer = pyproj.Transformer.from_crs(source_crs, target_crs, always_xy=True)

# Iterate over each pixel and retrieve its coordinates
with rasterio.open(image_path) as dataset:
    #height  = 4
    for i in range(4):
        for j in range(dataset.width):
            row, col = i, j
            x, y = dataset.xy(row, col)
            lon, lat = transformer.transform(x, y)
            print(f"Pixel ({row}, {col}): Longitude {lon}, Latitude {lat}")
```
